```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
haberman=pd.read_csv('haberman.csv')
```

# Violin Plot

```python
sns.violinplot(x='status',y='nodes',data=haberman)
```

```python
sns.violinplot(x='status',y='age',data=haberman)
```

```python
sns.violinplot(x='status',y='year',data=haberman)
```

```python
sns.violinplot(x='year',y='age',data=haberman)
```

```python
sns.violinplot(x='age',y='year',data=haberman)
```

# Box plot

```python
sns.boxplot(x='nodes',y='age',hue='status',data=haberman)
```

```python
sns.boxplot(x='status',y='age',data=haberman)
```

# Scatter plot

```python
sns.FacetGrid(data=haberman,hue='status').map(plt.scatter,'age','year').add_legend()
```

# Pair Plot

```python
sns.set_style('whitegrid')
sns.pairplot(data=haberman,hue='status', palette='Accent')
```

# Contour Plot

```python
sns.set_style('darkgrid')
sns.jointplot(x='status',y='age',data=haberman,kind='kde')
```

```python
sns.set_style('whitegrid')
sns.jointplot(x='age',y='year',data=haberman,kind='kde')
```
