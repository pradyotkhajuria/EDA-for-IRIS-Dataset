```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
#warnings.warn(msg, UserWarning)
```

```python
haberman=pd.read_csv('haberman.csv')
haberman
```

```python
sns.FacetGrid(data=haberman,hue='status',size=5).map(sns.distplot,'age').add_legend()
plt.show()
```

```python
sns.FacetGrid(data=haberman,hue='status',size=4).map(sns.distplot,'year').add_legend()
```

```python
sns.FacetGrid(data=haberman,hue='year',size=5).map(sns.distplot,'nodes').add_legend()
```
